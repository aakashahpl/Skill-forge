````python
from dotenv import load
import textwrap
import os
import re
import json
import google.generativeai as genai
from IPython.display import display
from IPython.display import Markdown
````

````python
def to_markdown(text):
    text = text.replace("â€¢", "  *")
    return Markdown(textwrap.indent(text, "> ", predicate=lambda _: True))
````

````python
load()
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
# GOOGLE_API_KEY = ("AIzaSyAo9ciCU1sXR9dUrQzyc17RBMeTV0dji6Q")
print(GOOGLE_API_KEY)

genai.configure(api_key=GOOGLE_API_KEY)
````

````python
for m in genai.list_models():
    if "generateContent" in m.supported_generation_methods:
        print(m.name)
````

````python
model = genai.GenerativeModel("gemini-1.0-pro-latest")
````

# COURSE GENERATION


````python
search = "I want to learn about python-basics"
language = "python"

response = model.generate_content(
    f"""
{search}.Make me a course on it. Suggest multiple modules (minimum 5 modules) on the Course along with the estimated time required to complete the course. If topic is a language, start from basics, else focus only on the topic in {language} language. Return the output in JSON in the following format.
Example output:
[
    {{
        "title": 
        "description": 
        "time-required": 
    }},
    {{
        "title": 
        "description": 
        "time-required": 
    }}
    # AND SO ON..
]
"""
)
````

````python
print(response)
raw_output = response.text
print("<------------------------------------------------------->")
print(type(raw_output))
print(raw_output)
````

````python
output_md = to_markdown(raw_output)
````

````python
output_md
````

````python
str_output = raw_output.replace("json", "").replace("JSON", "").strip("```")
print(str_output)
````

````python
json_data = json.loads(str_output)

# Now json_data is a Python object representing the JSON structure
print(json.dumps(json_data, indent=2))
````
